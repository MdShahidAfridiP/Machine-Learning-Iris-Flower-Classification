## IRIS dataset
### classification of Versicolor , Setosa , Virginica Flowers

![Classifiying flowers](iris.png)

```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df=pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
df.head()
```

```python
#univariate analysis
df['species'].unique()
df_setosa=df[df['species']=='setosa']
df_versicolor=df[df['species']=='versicolor']
df_virginica=df[df['species']=='virginica']

plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
plt.xlabel("sepal length")
```

```python
#bivariate analysis
sns.pairplot(x_vars='sepal_length',y_vars='sepal_width',hue='species',data=df,kind='scatter',palette='rainbow',size=4)
```

```python
#multivariate analysis
sns.pairplot(df,hue='species')
```

```python
x=df[['petal_width','petal_length']].values
y=df.iloc[:,-1].values
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)

#Random Forest

from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(xtrain,ytrain)

ypred=classifier.predict(xtest)

from sklearn.metrics import accuracy_score,confusion_matrix
result=accuracy_score(ytest,ypred)
print("accuracy : ",result*100)

result1=confusion_matrix(ytest,ypred)
print(result1)
```

# Logistic Regrssion

```python

from sklearn.linear_model import LogisticRegression
regression=LogisticRegression()
regression.fit(xtrain,ytrain)
ypred=regression.predict(xtest)

from sklearn.metrics import accuracy_score,confusion_matrix
result=accuracy_score(ytest,ypred)
print("accuracy : ",result*100)

result1=confusion_matrix(ytest,ypred)
print(result1)
```
